```python
import pandas as pd
data = pd.read_csv("data\\festival_data.csv", encoding='cp949')
data.head()
data = data.dropna()
```

```python
data["start"] = data["축제 기간"].apply(lambda x: int(x.split(" ~ ")[0].replace(".", "")))
data["end"] = data["축제 기간"].apply(lambda x: int(x.split(" ~ ")[1].replace(".", "")))

data.drop(columns = "축제 기간", inplace = True)
data.head()
```

```python
data["주소"] = data['주소'].apply(lambda x: " ".join(x.split()[:3]) if x.split()[0][-1] == "도" else " ".join(x.split()[:2]))
# data.drop(columns = "주소", inplace = True)
data.head()
```

```python
data["비용(상세)"] = data["비용"]
data["비용"] = data["비용"].apply(lambda x: x if x in ["무료", "유료"] else "부분유료")
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("intfloat/multilingual-e5-large-instruct")
data["embedding"] = data["상세 내용"].apply(lambda x: model.encode(x) if type(x) != float else model.encode("None"))
```

```python
print(data.head())
data = data.rename(columns={'제목': '축제이름', '상세 내용': '축제소개', 'start': '시작일', 'end': '마감일'})
print(data.head())
```

```python
import pickle
data["primary_key"] =  range(0, len(data))
with open('data/festival.pkl', 'wb') as file:
    pickle.dump(data, file)
```
